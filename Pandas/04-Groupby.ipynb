### **Groupby**

The `.groupby` allows you to group rows of data together and call aggregate functions.

```python
import numpy as np
import pandas as pd
```

```python
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

In using the `.groupby` method you can group rows together based off of a column name, in this instance we will use Company.

```python
byComp = df.groupby('Company')
```

Once it is saved as a new variable you can call aggregate methods off the object:

```python
byComp.std(numeric_only=True)
```

```python
byComp.sum(numeric_only=True).loc['FB']
```

```python
df.groupby('Company').min()
```

The `describe()` method returns description of the data in the DataFrame.

```python
df.groupby('Company').describe()
```

```python
byComp.describe().transpose()['GOOG']
```
