### **Operations**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'col1':[1,2,3,4],
                   'col2':[444,555,666,444],
                   'col3':['abc','def','ghi','xyz']})
df.head() 
```

To display the first 5 rows of a DataFrame use the method `.head()` 

## **Unique Values**

`.unique()` returns a NumPy array containing the unique values from a Pandas Series, in the order they appear.

`.nunique` returns the number of unique values for each column.

```python
df['col2'].nunique() # .unique() and .nunique()
```

```python
df['col2'].value_counts() 
```

## **Selecting Data**

```python
# Select from DataFrame using criteria from multiple columns
df[(df['col1']>2) & (df['col2']==444)]
```

## **Applying Functions**

```python
def times2(x):
    return x*2
```

```python
df['col1'].apply(times2)
```

```python
df['col2'].apply(lambda x: x*2)
```

```python
df.drop('col1',axis=1)
```

`.columns` returns the column labels of the DataFrame.

```python
df.columns
```

`.index` returns the index labels of the DataFrame.

```python
df.index
```

### **Sorting and Ordering a DataFrame**

`.sort_values(by, axis)` sort according to the defined parameters within the method.

```python
# inplace=False by default
df.sort_values(by='col2')
```

`.isnull()` finds or checks for null values.

```python
df.isnull()
```

```python
data = {'A':['foo','foo','foo','bar','bar','bar'],
     'B':['one','one','two','two','one','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]}

df = pd.DataFrame(data)
```
