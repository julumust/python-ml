### **Excel**

- Pandas can read and write excel files - it can only import data.
- It cannot import formulas, images or macros causing `read_excel()` to crash.

### Excel Input

```python
pd.read_excel('Excel_Sample.xlsx',sheet_name='Sheet1',index_col=0)
```

### Excel Output

```python
df.to_excel('Excel_Sample2.xlsx',sheet_name='NewSheet')
```

## **HTML**

- Pandas can read table tabs off of HTML.

### HTML Input

`read_html()` will read tables off of a webpage and return a list of DataFrame objects.

```python
data = pd.read_html('http://www.fdic.gov/bank/individual/failed/banklist.html')
```
